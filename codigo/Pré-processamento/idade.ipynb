```python
import pandas as pd
import numpy as np 


df = pd.read_pickle('C:/Users/Rafael/OneDrive/Documentos/TCC Rafaela/Dados/Pickle/cancer_mama_regiao_sudeste_com_desfecho.pickle')
```

```python
# Pivot da Tabela para pegar max e min da idade
pivot_table = pd.pivot_table(data=df,
                             index='AP_CNSPCN',
                             values=['AP_DTINIC', 'AP_NUIDADE'],
                             aggfunc=['min','max'])

#Renomeação das colunas
pivot_table = pivot_table.set_axis(['AP_DTINIC_min', 'idade_inicial', 'AP_DTINIC_max', 'idade_max'], axis=1)

#Merge com o dataset
df = pd.merge(df, pivot_table, on='AP_CNSPCN', how='left')
df.head()
```

```python
# Excluindo as colunas 
df.drop(columns=['AP_DTINIC_min',	'AP_DTINIC_max'], inplace=True)

df.head()
```

```python
def ajustar_idade(row):
    # Converter as colunas de string para datetime
    data_max = pd.to_datetime(row['fim'])
    data_inicial = pd.to_datetime(row['inicio'])

    # Calcular a diferença em anos
    anos_diff = (data_max - data_inicial).days // 365
    
    # Ajustar a idade se necessário (verificação de ano)
    if data_max < data_inicial.replace(year=data_max.year):
        anos_diff -= 1
    
    return anos_diff
```

```python

# Aplicando a função na DataFrame
df['idade_ajustada'] = df.apply(ajustar_idade, axis=1)
```
